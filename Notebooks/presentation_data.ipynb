```python
# Check missing values
print("Missing values in AQI dataset:\n", aqi.isnull().sum())
print("\nMissing values in Meteorological dataset:\n", met.isnull().sum())
print("\nMissing values in Fire dataset:\n", fire.isnull().sum())

# Basic statistics
print("\nAQI Statistics:\n", aqi.describe())
print("\nMeteorological Statistics:\n", met.describe())
print("\nFire Data Statistics:\n", fire.describe())
```
